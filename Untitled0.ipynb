```python
from flask import Flask, request, jsonify
from flask_cors import CORS
from sentence_transformers import SentenceTransformer, util

app = Flask(__name__)
CORS(app)

model = SentenceTransformer('all-MiniLM-L6-v2')

def evaluate_answer(user_answer, expected_answer):
    emb1 = model.encode(user_answer, convert_to_tensor=True)
    emb2 = model.encode(expected_answer, convert_to_tensor=True)

    similarity = util.pytorch_cos_sim(emb1, emb2)
    score = float(similarity[0][0]) * 100

    if score > 80:
        feedback = "Excellent answer! Very relevant."
    elif score > 60:
        feedback = "Good answer, but could be improved."
    else:
        feedback = "Answer needs improvement."

    return score, feedback

@app.route('/')
def home():
    return "API is running"

@app.route('/evaluate', methods=['POST'])
def evaluate():
    data = request.json
    user_answer = data['answer']
    expected_answer = data['expected']

    score, feedback = evaluate_answer(user_answer, expected_answer)

    return jsonify({
        "score": score,
        "feedback": feedback
    })
```
